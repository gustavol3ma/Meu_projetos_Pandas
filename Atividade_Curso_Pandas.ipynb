1) Os alunos participaram de uma atividade extracurricular e ganharam pontos extras. Esses pontos extras correspondem a 40% da nota atual de cada um deles. Com base nisso, crie uma coluna chamada "Pontos_extras" que contenha os pontos extras de cada aluno, ou seja, 40% da nota atual deles.

```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/desafios/alunos.csv'
dados_completo = pd.read_csv(url)
dados_completo
```

```python
dados_completo['Pontos_extras'] = dados_completo['Notas'] * 0.4
dados_completo
```

```python
dados_completo['Notas_Finais'] = dados_completo['Notas'] + dados_completo['Pontos_extras']
dados_completo
```

```python
dados_completo['Aprovação_Final'] = dados_completo['Notas_Finais'] >= 6
dados_completo
```

```python
dados_completo.query('Aprovado == False & Aprovação_Final == True')
selecao = (dados_completo['Aprovado'] == False) & (dados_completo['Aprovação_Final'] == True)
dados_completo[selecao]
```
